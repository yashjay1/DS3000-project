```python
import pandas as pd
```

```python
df = pd.read_csv('walmartData.csv')
# create index
df.rename(columns={'Unnamed: 0': '0'}, inplace = True)
df.set_index('0', inplace=True)
df.fillna(0, inplace=True)
# change the string review values to numeric floats
df['reviews'] = pd.to_numeric(df['reviews'].str.extract(r'(\d+)')[0], errors='coerce')
df.head(50)
```

```python
def create_classes(df):

    rank_changes = (df['rank'] - df['rank'].shift(1)) < 0

    # Create an array to hold class names
    class_names = []
    class_number = 1

    # Loop through the data to assign class names
    for i, change in enumerate(rank_changes):
        if change:
            class_number += 1
        class_names.append(f'{class_number}')

    # Add the 'Class' column to the DataFrame
    df['class'] = class_names
    
    return df
```

```python
df = create_classes(df)
df.head(50)
```
